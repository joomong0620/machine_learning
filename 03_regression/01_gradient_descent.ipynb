## 단순 선형 회귀(Simple Regression)를 통한 회귀의 이해


**회귀 모델 수식**


* **예측 모델**:T
 $f(x) = w_0 + w_1 \cdot x$


* 여기서:


 * $w\_0$: 절편 (bias)
 * $w\_1$: 기울기 (weight)
 * $x\_i$: 입력값 (주택 크기)
 * $y\_i$: 실제 주택 가격


**설명**


* 주택 가격은 주택 크기에 따라 선형적으로 결정된다고 가정.
* 모델은 예측값 $f(x)$를 생성.
* 실제값과 예측값 사이의 차이인 \*\*오차(Error)\*\*를 최소화하는 것이 목적.
* **최적의 회귀 모델**이란:


 * 전체 데이터의 **오차값 제곱의 합(RSS)** 이 **최소**가 되는 모델을 의미.
 * 즉, **오차 합이 최소가 되는 회귀계수 $w\_0$, $w\_1$** 을 찾는 것.




#### RSS 기반의 회귀 오차 측정


**RSS (Residual Sum of Squares)**


* 오차 제곱합을 더해서 모델의 성능을 평가하는 방식.
* **RSS 수식**:
 ```
 RSS = (y₁ - (w₀ + w₁ * x₁))²
     + (y₂ - (w₀ + w₁ * x₂))²
     + (y₃ - (w₀ + w₁ * x₃))²
     + ... (모든 학습 데이터에 대해 수행)
 ```


* 일반적으로 **미분을 통해 최적의 w₀, w₁ 값을 구할 때** 편리하게 쓰임.


**시각적 해석**


* 각 점마다 수직선은 **실제값과 예측값 사이의 오차**를 의미.
* 이 오차의 **제곱**을 모두 더한 것이 **RSS**이다.
* 이 값을 최소화하는 회귀선이 **최적의 선형 회귀 모델**이다.



# Gradient Descent


경사하강법(Gradient Descent)은 **최적의 값을 찾기 위해 손실 함수의 기울기(경사)를 따라 조금씩 이동하는 방법**이다.


<img src="https://d.pr/i/0vERpX+" alt="경사하강법" width="500">




| 항목 | 설명                                          |
| -- | ------------------------------------------- |
| 목적 | 손실 함수(Loss Function)를 최소화하는 가중치(파라미터)를 찾는 것 |
| 방식 | 기울기(미분 값)의 반대 방향으로 가중치를 업데이트                |
| 반복 | 일정 횟수(iter) 동안 또는 충분히 작아질 때까지 반복            |


**수식 형태**


* 손실 함수 $L(w)$에 대해
 **업데이트 식:**


 $$
 w := w - \eta \cdot \frac{dL(w)}{dw}
 $$


 * $w$: 파라미터 (가중치)
 * $\eta$: 학습률(learning rate)
 * $\frac{dL(w)}{dw}$: 손실 함수의 기울기 (미분값)






**주의할 점**


* **학습률**이 너무 크면 발산하고, 너무 작으면 느리게 수렴함
* **지역 최소값**에 빠질 수 있음 (비선형 함수일 경우)
* 다양한 변형 알고리즘이 존재함 (Momentum, Adam 등)


## RSS


회귀 분석에서 모델의 정확도를 평가하는 지표 중 하나는 **RSS**이다.
RSS는 "Residual Sum of Squares"의 약자로, **잔차 제곱합**을 의미한다.
모델의 예측값과 실제 관측값 간의 차이를 제곱한 후 모두 더한 값을 나타내며, 이 값이 작을수록 예측이 실제에 가까운 좋은 모델이라고 볼 수 있다.


RSS는 다음과 같이 계산한다:


$\text{RSS} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$


여기서:


* $y\_i$는 실제값을 의미한다.
* $\hat{y}\_i$는 예측값을 의미한다.
* $n$은 데이터의 개수이다.


RSS 값이 작다는 것은 모델의 예측이 실제 값에 가까움을 의미하며, 회귀 분석에서는 일반적으로 이 값을 최소화하는 방향으로 모델을 최적화한다.
이러한 RSS를 기반으로 만든 함수는 회귀에서 **비용 함수**, **손실 함수**, 혹은 **로스 함수**라고 부른다.


**비용 함수의 미분**


비용 함수 $R(w)$는 가중치 $w$에 대한 함수이며, RSS를 평균까지 포함하여 정의할 수 있다:


$\text{RSS} = \frac{1}{n}\sum_{i=1}^{n} (y_i - \hat{y}_i)^2$


회귀 모델의 목표는 이 비용 함수 $R(w)$의 값을 최소로 만드는 가중치 $w$를 찾는 것이다.
이를 위해 비용 함수를 **미분**하여 그 미분 함수의 최소값을 구해야 한다.


미분은 어떤 함수의 **순간 변화율**을 구하는 도구이다. 각 지점에서 기울기를 계산해 기울기가 0에 가까워질 때까지 $w$를 조정한다. 기울기의 부호를 통해 $w$를 줄일지, 늘릴지 판단할 수 있다.


하지만 실제로는 단순한 미분이 아닌 \*\*편미분(partial derivative)\*\*을 수행해야 한다.
이유는 비용 함수가 단일 변수 함수가 아니라, 절편 $w\_0$과 가중치 $w\_1$이라는 **두 개의 변수로 구성된 다변수 함수**이기 때문이다.


따라서:


* $w\_0$이 고정된 상태에서 $w\_1$에 대해 미분한 값을 계산하고,
* $w\_1$이 고정된 상태에서 $w\_0$에 대해 미분한 값을 계산한다.


이처럼 각 변수에 대해 부분적으로 미분을 수행하는 것을 **편미분**이라고 한다.
이 과정을 통해 비용 함수의 기울기를 구하고, 그 정보를 바탕으로 경사하강법 등의 최적화 알고리즘을 통해 최적의 파라미터 $w\_0$, $w\_1$을 찾아갈 수 있다.






### RSS의 미분
RSS(Residual Sum of Squares)는 회귀 분석에서 모델의 성능을 평가하는 데 사용되는 손실 함수이고, 이 함수를 최소화하는 것이 회귀 모델의 목표이다.


$
\mathrm{R}(\mathrm{w}) = \frac{1}{N} \sum_{i=1}^{N}\left(y_{i}-\left(w_{0}+w_{1} \cdot x_{i}\right)\right)^{2}
$


<br>


다음과 같이 z를 정의하면:


$
z = (y_i - (w_0 + w_1 \cdot x_i))
$


#### 1.$w_0$에 대한 편미분


다음과 같이 미분 연쇄법칙을 적용할 수 있다.


$
\frac{dR(w)}{dw_0} = \frac{dR(w)}{dz} \cdot \frac{dz}{dw_0}
$


위 식을 따로 계산해서 곱하면


1. **$\frac{dR(w)}{dz}$**:


   $
       \frac{dR(w)}{dz} = \frac{1}{N} \sum_{i=1}^{N} 2 z
   $


2. **$\frac{dz}{dw_0}$**:


   $
       \frac{dz}{dw_0} = -1
   $


3. **곱셈**:


   $
   \frac{dR(w)}{dw_0} = \frac{1}{N} \sum_{i=1}^{N} 2 z \cdot -1 = -\frac{2}{N} \sum_{i=1}^{N} (y_i - (w_0 + w_1 \cdot x_i))
   $


#### 2.$w_1$에 대한 편미분
다음과 같이 미분 연쇄법칙을 적용할 수 있다.


$
\frac{dR(w)}{dw_1} = \frac{dR(w)}{dz} \cdot \frac{dz}{dw_1}
$


위 식을 따로 계산해서 곱하면


1. **$\frac{dR(w)}{dz}$**:


   $
       \frac{dR(w)}{dz} = \frac{1}{N} \sum_{i=1}^{N} 2 z
   $


2. **$\frac{dz}{dw_1}$**:


   $
       \frac{dz}{dw_0} = -x_i
   $


3. **곱셈**:


   $
   \frac{dR(w)}{dw_1} = \frac{1}{N} \sum_{i=1}^{N} 2 z \cdot -x_i = -\frac{2}{N} \sum_{i=1}^{N} x_i \cdot (y_i - (w_0 + w_1 \cdot x_i))
   $


## 경사 하강법 구현

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error
````

````python
# 데이터 준비
np.random.seed(42) # 데이터 생성

# 입력데이터 생성(특성 하나)
# 0과 1 사이의 균등분포를 따르는 난수 반환
# -> 0 ~ 2 사이의 균등분포 난수
X = 2 * np.random.rand(100, 1)

# 타겟 데이터 생성
true_w = 6
true_b = 4
y = true_w * X + true_b

# 평균0, 표준편차1인 정규분포를 따르는 난수를 노이즈로 추가
y += np.random.randn(100, 1)

# 시각화
plt.scatter(X, y)
plt.show()
````

````python

# 가중치/절편 1번 업데이트
w0 = np.zeros((1,1)) # 절편(bias)
w1 = np.zeros((1,1)) # 기울기(weight)

# 예측 = w0 + w1 * 입력값
y_pred = w0 + w1 * X
#y_pred.shape

# 실제값 차이 계산
diff = y - y_pred
# print(diff)

# 데이터의 개수
N = len(X)

# 학습률 : 얼마나 움직일지 결정하는 크기
lr = 0.01


# w0 기울기 계산
# 기울기 = (-2/N) * 학습률 * diff 합
w0_grad = (-2/ N) * lr * np.sum(diff)
w0_grad

# w1 기울기 계산
# 기울기 = (-2/N) * 학습률 * (diff * X) 합
#                       (100, 1) (100, 1) -> 행렬곱셈 불가
w1_grad = (-2/ N) * lr * np.dot(diff.T , X) # (1, 100) (100, 1)
w1_grad


# 가중치/절편 업데이트
w0 -= w0_grad
w1 -= w1_grad # 손실이 줄어들었으면 좋겠다.

w0, w1
````

````python
# 경사하강법 구현 : 반복 처리를 통해 최적의 가중치/절편을 찾는다.

def gradient_descent(X, y, n_iter=1000, lr=0.01) :
    w0 = np.zeros((1, 1)) # 절편(bias)
    w1 = np.zeros((1, 1)) # 기울기(weight)
    N = len(X) # 데이터의 개수

    for i in range(n_iter):

        # 예측 = w0 + w1 * 입력값
        y_pred = w0 + w1 * X

        # 실제값 차이 계산
        diff = y - y_pred

        # w0 기울기 계산
        # 기울기 = (-2/N) * 학습률 * diff합
        w0_grad = (-2 / N) * lr * np.sum(diff)

        # w1 기울기 계산
        # 기울기 = (-2/N) * 학습률 * (diff * X)합
        #                       (100,1) (100,1) -> 행렬곱셈 불가
        w1_grad = (-2 / N) * lr * np.dot(diff.T, X) # (1, 100) (100,1)

        # 가중치/절편 업데이트
        w0 -= w0_grad
        w1 -= w1_grad

        # 시각화
        if i % 50 == 0 :
            plt.figure(figsize=(3,2))
            plt.scatter(X, y)
            plt.plot(X, w0[0, 0] + w1[0, 0] * X, color='r')
            plt.title(f'Step {i}')
            plt.show()

    return w0, w1


# true=b =4, true_w = 6
w0, w1 = gradient_descent(X, y, n_iter=1000)
w0, w1

````

## 미니배치 경사하강법
- 기존방식: 한번에 모든 X데이터에 대해 예측/기울기계산/반영 작업 진행(epoch) 지정한 n_iter만큼 반복
- 미니배치 경사하강법: 한번의 epoch에 지정한 개수의 mini-batch만 샘플링해서 진행
- 확률적 경사하강법(SGD): mini-batch크기가 1인 방식


````python
def mini_batch_gradient_descent(X, y, batch_size=10, n_iter=1000, lr=0.01):
    w0 = np.zeros((1, 1))
    w1 = np.zeros((1, 1))
    N = len(X)

    # 한번 epoch에 batch_size만큼 데이터를 샘플링해서 진행
    for i in range(n_iter) :
        indices = np.random.permutation(N)[:batch_size]
        # N개의 데이터를 섞어서  batch_size만큼 인덱스 반환

        X_batch = X[indices]
        y_batch = y[indices]

        # 예측 = w0 + w1 * 입력값
        y_pred = w0 + w1 * X_batch

        # 실제값 차이 계산
        diff = y_batch - y_pred

        # 각각 기울기 계산
        # 기울기 = (-2/N) * 학습률 * diff합
        w0_grad = (-2 / N) * lr * np.sum(diff)
        w1_grad = (-2 / N) * lr * np.dot(diff.T, X_batch)
        # 가중치/절편 업데이트
        w0 -= w0_grad
        w1 -= w1_grad

        # 시각화
        if i % 50 == 0 :
            plt.figure(figsize=(3,2))
            plt.scatter(X, y)
            plt.scatter(X_batch, y_batch)
            plt.plot(X_batch, w0[0, 0] + w1[0, 0] * X_batch, color='r')
            plt.title(f'Step {i}')
            plt.show()
    return w0, w1



w0, w1 = mini_batch_gradient_descent(X, y)
w0, w1
````

## 확률적 경사하강법 Stochastic Gradient Descent

````python
def stochastic_gradient_descent(X, y, n_iter=5000, lr=0.01):
    w0 = np.zeros((1, 1))
    w1 = np.zeros((1, 1))
    N = len(X)

    # 한번 epoch에 batch_size만큼 데이터를 샘플링해서 진행
    for i in range(n_iter) :
        index = np.random.randint(0, N)

        X_i = X[index:index + 1] # (1,1) 차원 유지를 위해서 슬라이싱 해줌
        y_i = y[index:index + 1] # (1,1) 차원 유지를 위해서 슬라이싱 해줌

        # 예측 = w0 + w1 * 입력값
        y_pred = w0 + w1 * X_i

        # 실제값 차이 계산
        diff = y_i - y_pred

        # 각각 기울기 계산
        # 기울기 = (-2/N) * 학습률 * diff합
        w0_grad = (-2 / N) * lr * np.sum(diff)
        w1_grad = (-2 / N) * lr * np.dot(diff.T, X_i)
        # 가중치/절편 업데이트
        w0 -= w0_grad
        w1 -= w1_grad

        # 시각화
        if i % 500 == 0 :
            plt.figure(figsize=(3,2))
            plt.scatter(X_i, y_i)
            plt.plot(X, w0[0, 0] + w1[0, 0] * X, color='r')
            plt.title(f'Step {i}')
            plt.show()
    return w0, w1



w0, w1 = stochastic_gradient_descent(X, y)
w0, w1
````
